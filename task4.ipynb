```python
import numpy as np

def prewitt_h(img, size):
    result = np.zeros_like(img)
    n = (size - 1) // 2

    for p in np.ndindex(img.shape[0]):
        neighborhood_sum = 0

        for q in np.ndindex(size, size):
            if p[0] + q[0] - n >= 0 and p[0] + q[0] - n < img.shape[0]:
                neighborhood_sum += img[p[0] + q[0] - n, p[1]] * (q[0] - n)

        result[p] = neighborhood_sum * (1 / 6) if 0 <= p[0] < img.shape[0] else 0

    return result
```

```python
lena_prewitt_h = img.copy()
plt.imshow(prewitt_h(lena_prewitt_h,3))
plt.colorbar()
```

```python
def prewitt_v(img, size):
    # write your code here
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            # write your code here
            if p[1]>=n and p[1]<img.shape[1]-n: 
                if q[1]==0:
                    neighborhood_sum += img[(p[1]+ q[1]-n)]
                elif q[1]==1:
                    neighborhood_sum += img[(p[1]+ q[1]-n)]*0
                elif q[1]==2:
                    neighborhood_sum += img[(p[1]+ q[1]-n)]*(-1)
                result[p] = neighborhood_sum*(1/6)
            else: 
                result[p]=0    

    return result 
```

```python
lena_prewitt_v = img.copy()
plt.imshow(prewitt_v(lena_prewitt_v,3))
plt.colorbar()
```
